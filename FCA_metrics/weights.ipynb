```python
import numpy as np
import matplotlib.pyplot as plt
def f(x,beta):
    return np.exp(-x*x/beta)

def poids(x,beta,cach_area):
    d=np.arange(0,cach_area+1,5)
    n=len(d)
    for i in range(n-1):
        if x>=d[i] and x<d[i+1]:
            return f(d[i],beta)
    return 0

def weight_discret(l_beta,cach_area):
    d=np.linspace(0,cach_area,10000)
    n=len(l_beta)
    T=[]
    for i in range(n):
        p=[poids(x,l_beta[i],cach_area) for x in d]
        T.append(p)
        plt.scatter(d,T[i],marker="_",label="\u03B2="+str(l_beta[i]))
    plt.xlabel("Travel cost")
    plt.ylabel("Value of weight")
    plt.legend()
    plt.show()

weight_discret([35,100],35)







```

```python
def weight_continue(l_beta,cach_area):
    n=len(l_beta)
    d=np.linspace(0,cach_area+10,1000)
    T=[]
    for i in range(n):
        l=f(d,l_beta[i])
        T.append(l)
        plt.plot(d,T[i],label="\u03B2="+str(l_beta[i]))
    plt.xlabel("Travel cost")
    plt.ylabel("Value of weight")
    plt.legend()
    plt.show()

weight_continue([266,366,466,566,666,766,866,966,1066],35)
```

```python

def comparaison(beta,cach_area):
    d=np.linspace(0,cach_area+10,10000)
    poids1=[1 if x<cach_area else 0 for x in d]
    poids2=[poids(x,beta,cach_area) for x in d]
    poids3=[f(x,beta) if x<35 else 0 for x in d]
    plt.plot(d,poids2, label="E2SFCA")
    plt.plot(d,poids3, label="KD2SFCA")
    plt.plot(d,poids1,"b--",marker="_",label="2SFCA")
    plt.legend()
    plt.show()

comparaison(390,35)
```
